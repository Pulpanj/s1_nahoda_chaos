## Par zakladnich pravidel jak spoustet programy v Google COLAB

- program se zde zapisuje do bunek (nekolik radek kodu), bunka kodu se prida pomoci horniho menu (**+Code**), bunky lze posouvat (po kliknuti se objevi sipky nahoru a dolu, kterymi se bunka posouva)

- program v bunce se spusti kliknutim na trojuhelnik vlevo (na okraji bunky) nebo **Ctrl Enter** (stisknuti 2 klaves najednou), pokud bunka generuje nejaky vystup zobrazi se pod bunkou

- lze take vybrat jen cast bunky (pomoci mysi) a po stisku praveho tlacika mysi se objevi volba **Run Selection** - nebo lepe pomoci klavesove zkratky **Shift Ctrl Enter** (stisknuti 3 klaves najednou)


- vystup se generuje bud prikazem **print** (nebo v pripade grafu prikazy **plot**)

- pokud bunka na konci obsahuje jen promennou pak se vystup tak zobrazi hodnota  

- jednotlive bunky lze spoustet samostatne v jakekoliv poradi, stav promenych se "pamatuje" mezi bunkami (pozor pokud napr. nejaka promenna neni definovana a chceme ji pouzit v jine bunce tak dojde k chybe)

- bezpecne overeni ze vse je spravne je spustit vsechny bunky pomoci horniho menu (**|>Run All**)  

- pokud si nejsme jisti tak vedle (**|>Run All**) je sipka dolu, ktery rozbali menu obsahujici (**Restart Session**), jehoz vyber vymaze vsechny promenne a nastavi cely program do pocatecniho stavu

-  menu otevirane sipkou dolu vedle (**|>Run All**)  take obsahuje  (**Clear All Outputs**) ktere smaze vystupy vsech bunek

```python
# podmineny prikaz (porovnani >,<,>=,=<,==,!=)
K=8
if K >= 2:
  print("K je vetsi nebo rovno nez 2")
else:
  print("K je mensi nez 2")
```

```python
# Prikaz if vyhodnocuje logickou podminku - hodnotu true nebo false (pravda nepravda)
#  otestuj pomoci vyberu radek oddelenych prazdnym radkem a stiskem Shift Ctrl Enter

a = True    # pravda
b = False   # nepravda
c = a or b  # nebo -> True
c

d = a and b # a zároveň -> False
d

c = not a   # negace -> False
c
```

```python
# porovnani vraci logickou hodnotu (https://www.umimeinformatiku.cz/programovani-logicke-vyrazy)
#  otestuj pomoci vyberu radek oddelenych prazdnym radkem a stiskem Shift Ctrl Enter

x = 42     # přiřazení do proměnné
x == 42    # test rovnosti

x != 42    # test nerovnosti

x > 42     # je větší

x >= 42    # je větší nebo rovno
```

```python
# cyklus - tiskni cisla od 0 do 5 a jejich druhe mocniny
NMC=5
for i in range(1,NMC+1):      # range = rozsah od 1 az dokud i neni rovno NMC+1
  i2=i*i
  r01=random()
  print("i=",i,"i2=",i2,"random cislo 0..1:",r01)
```

```python
# podmineny prikaz v cyklu
for i in range(1,NMC+1):      # range = rozsah od 1 az dokud i neni rovno NMC+1
    r01=random()
    print("i=",i,"random cislo 0..1:",r01)
    if (i % 2) == 0:        # test zdali i je sude cislo ( zbytek po deleni i dvemi je 0)
        print("i je sude")
    else:
        print("K je liche")
print("konec cyklu")
```

```python
#malovani grafu

import numpy  # tohle je nutno dat nekam na zacatek programu, aby python zacal znat pole a vektory
import matplotlib.pyplot as plt # tohle je nutno dat nekam na zacatek programu, aby python zacal znat malovani grafu

# nadefinuj pole souradnic
x=numpy.zeros(5)
y2=numpy.zeros(5)
y3=numpy.zeros(5)

# x=numpy.array([1,2,3,4,5,])  # lze napsat rucne jako pole definovane seznamem
# y=numpy.array([1,4,9,16,25])  # lze napsat rucne jako pole definovane seznamem
for i in range(5):
  x[i]=i
  y2[i]=i*i
  y3[i]=i**3

# nadefinuj graf
plt.figure(figsize=(8, 6))      #velikost grafu v palcich
plt.plot(x, y2, label='(x**2)', color='blue',marker='x') # prvni cara
plt.plot(x, y3, label='(x**3)', color='red',marker='o') # druha cara
plt.title('Graf mocnin')
plt.xlabel('x')
plt.ylabel('Y')
plt.grid(True)      # pridej mrizku
plt.legend()      # pridej popisku car
plt.show()      # zobraz nadefinovane
```
